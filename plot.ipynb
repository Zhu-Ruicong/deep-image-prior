```python
import pandas as pd
```

```python
dip_file = "/Users/ruicong/Documents/exp_results/scalar/results_with_dip.csv"
softplus_10000_file = "/Users/ruicong/Documents/exp_results/scalar/results_softplus_with_scalar_10000.csv"
dip = pd.read_csv(dip_file)
softplus_10000 = pd.read_csv(softplus_10000_file)
dip = dip.rename(columns={"psrn_noisy": "dip_psnr_noisy", "psrn_gt": "dip_psnr_gt", "psrn_gt_sm": "dip_psnr_gt_sm"})
softplus_10000 = softplus_10000.rename(columns={"psrn_noisy": "softplus_10000_psnr_noisy", "psrn_gt": "softplus_10000_psnr_gt", "psrn_gt_sm": "softplus_10000_psnr_gt_sm"})
```

```python
merged = pd.merge(dip, softplus_10000, on=["iteration"])
```
